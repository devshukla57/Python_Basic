```python
#checking the data through groupby that how many male and female servived
```

```python
survival_counts= df.groupby(['sex', 'survived']).size().unstack()
```

```python
survival_counts
```

```python
#rename the columns for clarity
survival_counts.columns = ['Did not survive', 'survived']
print(survival_counts)
```

```python
#outliers
sns.boxplot(data=df, y='age')
plt.ylabel('Age')
plt.title('Box plot of Age')
plt.show()
```

```python
#checking outlier by using histogram
sns.histplot(data=df, x='age', bins=20, kde=True)
```

```python
#by using scatter plot
plt.scatter(df['pclass'], df['survived'])
```

```python
#Data Imbalance
print(df['survived'].value_counts())
```

```python
#correlation Analysis
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
